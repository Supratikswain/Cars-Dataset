```python
# Load the cars dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings(action="ignore", module="scipy", message="^internal gelsd")

df = pd.read_csv('cars.csv')
df.head(20)
df
```

```python
# Check for NaN values
df.isnull().any()
```

```python
# Fill the NaN values

from statistics import mode
df['gearbox']=df['gearbox'].fillna(mode(df['gearbox']))
df['model']=df['model'].fillna(mode(df['model']))
df['fuelType']=df['fuelType'].fillna(mode(df['fuelType']))
df['vehicleType']=df['vehicleType'].fillna(mode(df['vehicleType']))
df.head(30)
```

```python
df.replace({'gearbox':{'manuell':'manual','automatik':'automatic'}},inplace=True)
df.replace({'vehicleType':{'kleinwagen':'small_car','kombi':'combi','andere':'Others'}},inplace=True)
df.replace({'fuelType':{'benzin':'petrol','andere':'others','elektro':'electro'}},inplace=True)
df
```

```python
#Data Cleaning or formatting

#df.name=df.name.str.replace('_',' ')
df.model=df.model.str.replace('_',' ')
df.brand=df.brand.str.replace('_',' ')
df.head(25)
```

```python
#df.drop('model',axis=1,inplace=True)
#df.drop('brand',axis=1,inplace=True)
df.drop('name',axis=1,inplace=True)
#df.drop('monthOfRegistration',axis=1,inplace=True)
#df.drop('powerPS',axis=1,inplace=True)
#df.drop('model',axis=1,inplace=True)
#df.drop('kilometer',axis=1,inplace=True)
df.head()
```

```python
df = df.loc[(df.price>400)&(df.price<=40000)]
df = df.loc[(df.yearOfRegistration>1990)&(df.yearOfRegistration<=2016)]
df = df.loc[(df.powerPS>10)]
df = df.loc[(df.kilometer>1000)&(df.kilometer<=150000)]
df
```

```python
#Replacing all the 0 month values to 1
df.monthOfRegistration.replace(0,1,inplace=True)
# Making the year and month column to get a single date
Purchase_Datetime=pd.to_datetime(df.yearOfRegistration*10000+df.monthOfRegistration*100+1,format='%Y%m%d')
import time
from datetime import date
import datetime as dt
y=dt.datetime(2018, 11,1)
# Calculating days old by subracting both date fields and converting them into integer
Days_old=(y-Purchase_Datetime)
Days_old=(Days_old / np.timedelta64(1, 'D')).astype(int)
#type(Days_old[1])
df['Days_old']=Days_old
df
```

```python
df.drop(columns=['yearOfRegistration','monthOfRegistration','powerPS'],inplace=True)
df
```

```python
#cat_cols = ['gearbox','fuelType','brand','vehicleType']
#df = pd.get_dummies(df, columns = cat_cols, drop_first=True)

df_dummies=pd.get_dummies(data=df,columns=['vehicleType','model','brand','gearbox','fuelType'],drop_first=True)
df_dummies.head()
```

```python
X = df_dummies.drop('price',axis=1)
X.head()
```

```python
y = df_dummies.price
y.head()
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X,y)

```

```python
print (linreg.intercept_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.20, random_state=0)
linreg = LinearRegression()
linreg.fit(X_train, y_train)
y_pred = linreg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

#rmse = sqrt(mean_squared_error(y_test, y_pred))
rmse=np.sqrt(metrics.mean_squared_error(y_test, y_pred))
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

adj_r2 = 1 - float(len(y)-1)/(len(y)-len(linreg.coef_)-1)*(1 - r2)

#rmse

#y_pred = linreg.predict(X_test)
#print(linreg.score(X_test, y_test)*100,'% Prediction Accuracy')
rmse, r2, adj_r2
```
